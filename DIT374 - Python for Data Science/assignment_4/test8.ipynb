```python
def merge_k_sorted_lists(lst):
    """
    input: 
    - lst: a list with lists 
    
    this fuction takes in a list and calculates the length of the input list
    and the lengt och the sub-lists inside
    
    OBS! all sub-lists are the same length
    """
    k = len(lst)
    n = len(lst[0])
    res_lst = [0] * (n * k); 
    splitter(0, k - 1, res_lst, lst, n, k)

    for i in range(n * k) :
        print(res_lst[i], end = " ");
```

```python
def splitter(left, right, res_lst, lst, n, k):
    """
    input:
    - left: int
    - right: int
    - res_lst: list with 0's of the length of arr
    - lst: is the list with list
    - n: int
    
    splitter creates adds the values of lst to res_lst.
    then splitter calls it selfe to split the list and calulate 
    the left & right values  
    """
    
    # if left & right are the same length add a new list with length n
    if left == right:
        for i in range(n):
            res_lst[left * n + i] = lst[left][i]
        return
    k = len(lst)
    # recursive call
    splitter(left, (left + right) // 2, res_lst, lst, n, k)
    splitter((left + right) // 2 + 1, right, res_lst, lst, n, k)
    merge(left, right,res_lst, n, k)
    
```

```python
def merge(left, right, result, n, k):
    """
    input:
    - left
    - right
    - result
    - n
    
    this method takes the list with values and and splits them to
    left and right start values. Then we get the length values of left and right.
    
    after we got the values we move over values to left and right by looping,
    using the length-values calculated before. 
    
    Lastly, this method loop throught left and right 
    """
    # store the initialize satr
    left_start = left * n
    right_start = ((right + left) // 2+1) *n
    
    # gives the size of left & right
    left_current = ((left + right) // 2 + 1) * n
    right_current = (right - (left + right) // 2) * n
    
    # list to work with left & right by using the size of the current left and right
    left_temp = [0] * left_current
    right_temp = [0] * right_current
    
    # moving left and right vaues to individual lists
    for i in range(left_current):
        left_temp[i] = result[left_start + i]
    for i in range(right_current):
        right_temp[i] = result[right_start + i]
    
    i = 0 # index for left_current
    j = 0 # index for right_current
    k = 0 # index for result
    
    print('CURRENT: ', left_current, right_current)
    print(n)
    print(k)
    #print(left_temp, right_temp)
    
    while i < left_current and j < right_current:
        if left_temp[i] < right_temp[j]:
            result[k] = left_temp[i]
            print('result', result)
            print('left_temp', left_temp)
            print('i:',i,'j:', j, 'k:', k)
            print()
            i += 1
        else:
            result[k] = right_temp[j]
            j += 1
            print('result', result)
            print('right_temp', right_temp)
            print('i:',i,'j:', j, 'k:', k)
            print()
        k += 1
    
    while i < left_current:
        result[k] = left_temp[i]
        print('result', result)
        print()
        i += 1
        k += 1
    
    #print(i, j, k)    
    #print('J R_C', j, right_current)
    while j < right_current:
        result[k] = right_temp[j]
        print('result', result)
        print()
        j += 1
        k += 1

    
    #print(i, j, k)
    return result

    
```
