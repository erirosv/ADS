```python
#Libraries
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 18
```

```python
data =[14, 7, 5, 4]
freq_series = pd.Series(data)
labels = ['Filter', 'Embedded', 'Wrapper', 'Hybrid']
```

```python
plt.figure(figsize=(12, 8))
ax = freq_series.plot(kind="bar")
ax.set_title("FS approach", fontsize=25)
ax.set_xlabel("Individual FS approach", fontsize=22)
ax.set_ylabel("Amount of individual FS approach", fontsize=22)
plt.xticks(rotation=45)
ax.set_xticklabels(labels)
```
