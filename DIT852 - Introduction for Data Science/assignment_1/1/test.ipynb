```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
data = pd.read_csv("food-supply-kcal.csv")
data.plot()
```

```python
x = []
y = []

start = 0
end = 1000
       
y = data['Year']
x = data['Food supply (kcal/capita/day) (FAO, 2020)']
plt.ylim(1960, 2020)
plt.xlim(1400, 4000)

#plt.xscale('log')
#plt.yscale('log')

plt.xlabel("Food supply (kcal)")
plt.ylabel("Year")

plt.xticks(rotation = 25)
plt.scatter(x[start:end],y[start:end])
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x = data['Food supply (kcal/capita/day) (FAO, 2020)'], y = data['Year'])
plt.xlabel("Food supply (kcal)")
plt.ylabel("Year")
plt.title('food supply (kcal)', fontsize = 20)
plt.ylim(1960, 2020)
plt.xlim(1200, 4000)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(y = data['Food supply (kcal/capita/day) (FAO, 2020)'], x = data['Entity'])
plt.ylabel("Food supply (kcal)")
plt.xticks(rotation = 25)
plt.xlabel("Countries")
plt.title('food supply (kcal)', fontsize = 20)
plt.xlim(10, 20)
plt.ylim(1000, 4000)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x = data['Year'], y = data['Food supply (kcal/capita/day) (FAO, 2020)'])
plt.ylabel("Food supply")
plt.xlabel("Year")
plt.title('food supply kcal', fontsize = 20)
plt.xlim(1960, 2020)
plt.ylim(1200, 4000)
plt.show()
```

```python
Names = []
Code = []
Year = []
Supply = []

with open('food-supply-kcal.csv','r') as csvfile:
    lines = csv.reader(csvfile, delimiter=',')
    for row in lines:
        Supply.append(row[0])
        #Code.append(row[1])
        Year.append(row[1])
        #Supply.append(row[3])

plt.scatter(Supply, Year, s = 500)
plt.xticks(rotation = 25)
plt.title('food supply kcal', fontsize = 20)
plt.xlabel("Supply")
plt.ylabel("Year")
plt.show()

#data.plot(kind='scatter',x='x',y='y')

#data.plot(plt.scatter(x,y))
```
