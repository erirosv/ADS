```python
import pandas as pd
import matplotlib as plt
```

```python
df_life = pd.read_csv('life-expectancy.csv')
df_life
```

```python
df_gdp = pd.read_csv('gdp-per-capita.csv')
df_gdp
```

```python
df = pd.merge(
    df_life,
    df_gdp,
    on = ['Entity', 'Year', 'Code'],
)
df=df.drop('145446-annotations', axis=1)
df=df[df['Year']>=1970]
df=df[df['Entity']!='World']
df['GDP per capita'].replace([0], 23906, inplace=True) # replaced missing value for the UAE with the mean of the two adjacent years 
df
```
