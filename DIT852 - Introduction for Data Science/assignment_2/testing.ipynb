### Problem 1

```python
# libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# read csv-file
data = pd.read_csv('data_assignment2.csv')
```

```python
# checking that the data is loaded correctly
data.plot.scatter('Living_area', 'Selling_price')
```

### 1-a

```python
x = []
y = []

x = data['Living_area']
y = data['Selling_price']

model = LinearRegression(normalize=True)
model.fit(x[:, np.newaxis], y) # f(x) = kx + m - regression line

x_shape = np.array([0,200])
y_shape= model.predict(x_shape[:,np.newaxis])

plt.scatter(x,y)
plt.plot(x_shape, y_shape)

plt.show()
```

### 1-b

```python
print('intercept: ', model.intercept_)
print('coeficent: ', model.coef_)
```
